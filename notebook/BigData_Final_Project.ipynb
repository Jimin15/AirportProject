```python
import pandas as pd

# 병합할 파일 경로 리스트
file_paths = [
    "Detailed_Statistics_Departures_SAN_origin.csv",
    "Detailed_Statistics_Departures_LGA_origin.csv",
    "Detailed_Statistics_Departures_LAX_origin.csv",
    "Detailed_Statistics_Departures_LAS_origin.csv",
    "Detailed_Statistics_Departures_EWR_origin.csv"
]

# 첫 번째 파일에서 스키마(헤더 포함) 읽기
try:
    merged_data = pd.read_csv(file_paths[0])
except FileNotFoundError:
    print(f"File not found: {file_paths[0]}")
    raise

# 나머지 파일들을 병합
for file_path in file_paths[1:]:
    try:
        data = pd.read_csv(file_path, skiprows=1)  # 1번째 행부터 읽기
        data.columns = merged_data.columns  # 열 이름을 첫 번째 파일과 동일하게 설정
        merged_data = pd.concat([merged_data, data], ignore_index=True)
    except FileNotFoundError:
        print(f"File not found: {file_path}")
        raise

# 최종 데이터 CSV로 저장
merged_data.to_csv("Detailed_Statistics_Departures_merged.csv", index=False)

print("모든 파일이 성공적으로 병합되었습니다!")
```

```python
import pandas as pd

# 데이터 읽기
data = pd.read_csv("Detailed_Statistics_Departures_merged.csv")

# 1. 날짜 형식 변환
data['Date'] = pd.to_datetime(data['Date (MM/DD/YYYY)'], format='%m/%d/%Y', errors='coerce')

# 2. 요일 정보 추가 (0=월요일, 6=일요일)
data['DayOfWeek'] = data['Date'].dt.dayofweek

# 3. 지연 여부 추가
data['IsDelayed'] = (data['Departure delay (Minutes)'] > 0).astype(int)

# 4. 총 지연 시간 추가
delay_columns = [
    'Delay Carrier (Minutes)',
    'Delay Weather (Minutes)',
    'Delay National Aviation System (Minutes)',
    'Delay Security (Minutes)',
    'Delay Late Aircraft Arrival (Minutes)'
]
data['Total Delay'] = data[delay_columns].sum(axis=1)

# 변환 결과 저장
data.to_csv("Detailed_Statistics_Departures_transformed.csv", index=False)

# 데이터 확인
print(data.head())
```

```python
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 읽기
data = pd.read_csv("Detailed_Statistics_Departures_transformed_cleaned.csv")

# 1. 요일별 지연 시간 (Departure delay) 계산
# 요일별 Departure delay (Minutes) 값 그룹화
delayed_times_by_weekday = [data[data['DayOfWeek'] == day]['IsDelayed'] for day in range(7)]

# 2. ANOVA 분석
f_stat, p_value = stats.f_oneway(*delayed_times_by_weekday)

# 3. 결과 해석
if p_value < 0.05:
    print("요일별 비행기 출발 지연 시간이 유의미하게 다릅니다.")
else:
    print("요일별 비행기 출발 지연 시간이 유의미한 차이가 없습니다.")

# 4. 요일별 지연 시간 박스 플롯 시각화
plt.figure(figsize=(10,6))
sns.boxplot(x='DayOfWeek', y='Departure delay (Minutes)', data=data)
plt.title('요일별 비행기 출발 지연 시간 분포')
plt.xlabel('요일')
plt.ylabel('출발 지연 시간 (분)')
plt.xticks(ticks=range(7), labels=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import f_oneway
import seaborn as sns

# 데이터 로드
data = pd.read_csv("Detailed_Statistics_Departures_transformed_cleaned.csv")

# 결측치 제거 (필요시)
data = data.dropna(subset=['IsDelayed', 'DayOfWeek'])

# 요일별 지연율 계산: 지연된 항공편 수 / 총 항공편 수
delay_rate_by_day = data.groupby('DayOfWeek')['IsDelayed'].mean()

# ANOVA 검정
delays = [data[data['DayOfWeek'] == day]['IsDelayed'] for day in data['DayOfWeek'].unique()]
f_stat, p_value = f_oneway(*delays)

# 결과 해석
if p_value < 0.05:
    anova_result = "요일별 지연율의 차이가 통계적으로 유의미합니다."
else:
    anova_result = "요일별 지연율의 차이가 통계적으로 유의미하지 않습니다."

# 3. ANOVA 결과 시각화 (F-statistic과 p-value 표시)
plt.figure(figsize=(8, 6))
sns.boxplot(x='DayOfWeek', y='IsDelayed', data=data, palette="viridis")
plt.axhline(y=delay_rate_by_day.mean(), color='red', linestyle='--', label=f'Overall Mean: {delay_rate_by_day.mean():.2f}')
plt.xlabel('Day of the Week')
plt.ylabel('Delay (Minutes)')
plt.title(f'ANOVA Test Result: {anova_result}')
plt.legend()
plt.show()

# ANOVA 검정 결과 출력
print(f"ANOVA test: F-statistic = {f_stat}, p-value = {p_value}")
print(f"결과 해석: {anova_result}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 로드
file_path = "Detailed_Statistics_Departures_transformed_cleaned.csv"
data = pd.read_csv(file_path)

# 결측치 제거 및 필요 없는 컬럼 제거
data = data.dropna(subset=['IsDelayed', 'DayOfWeek', 'Carrier Code', 'Destination Airport', 'Origin Airport',
                           'Delay Carrier (Minutes)', 'Delay Weather (Minutes)',
                           'Delay National Aviation System (Minutes)', 'Delay Security (Minutes)',
                           'Delay Late Aircraft Arrival (Minutes)'])

# 1. 전체 지연 원인 비율 계산
# 지연 원인 컬럼
cause_columns = ['Delay Carrier (Minutes)', 'Delay Weather (Minutes)', 
                 'Delay National Aviation System (Minutes)', 'Delay Security (Minutes)', 
                 'Delay Late Aircraft Arrival (Minutes)']

# 각 지연 원인의 합계 계산
total_delays = data[cause_columns].sum()

# 전체 지연 시간 중 각 원인이 차지하는 비율 계산
total_delay_ratios = (total_delays / total_delays.sum()) * 100

# 결과 출력
print("전체 지연 원인이 차지하는 비율 (%):")
print(total_delay_ratios)

# 2. 지연 원인 비율 시각화 (Pie Chart)
plt.figure(figsize=(8, 8))
plt.pie(total_delay_ratios, labels=total_delay_ratios.index, autopct="%.1f%%", startangle=140, colors=sns.color_palette("pastel"))
plt.title('Proportion of Overall Delay Causes')
plt.show()

# 3. 요일별 지연 원인 비율 계산 및 시각화
# 요일별 지연 원인의 합계 계산
delay_by_day = data.groupby('DayOfWeek')[cause_columns].sum()

# 요일별 총 지연 시간 추가
delay_by_day['Total Delay'] = delay_by_day.sum(axis=1)

# 요일별 지연 원인 비율 계산
delay_ratios = delay_by_day.div(delay_by_day['Total Delay'], axis=0) * 100
delay_ratios = delay_ratios.drop(columns='Total Delay')

# 요일별 지연 원인 비율 시각화 (Stacked Bar Chart)
ax = delay_ratios.plot(kind='bar', stacked=True, figsize=(10, 6), cmap="viridis")
plt.title('Proportion of Delay Causes by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Proportion of Delay (%)')
plt.legend(title='Delay Cause')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 로드
file_path = "Detailed_Statistics_Departures_transformed_cleaned.csv"
data = pd.read_csv(file_path)

# 결측치 제거 및 필요 없는 컬럼 제거
data = data.dropna(subset=['IsDelayed', 'DayOfWeek', 'Destination Airport', 'Origin Airport', 'Total Delay'])

# 2. 요일별 출발-도착 노선 평균 지연 분석
# 출발-도착 공항 쌍 및 요일별 평균 지연 시간 계산
origin_dest_day_delay = data.groupby(['Origin Airport', 'Destination Airport', 'DayOfWeek'])['Total Delay'].mean().reset_index()

# 데이터 피벗팅
pivot_data = origin_dest_day_delay.pivot_table(index=['Origin Airport', 'Destination Airport'], columns='DayOfWeek', values='Total Delay')

# 상위 7개 출발-도착 노선 추출
top7_routes = pivot_data.mean(axis=1).sort_values(ascending=False).head(7)

# 상위 7개 노선 값 출력
print("Top 7 Origin-Destination Routes by Average Delay Across Days (Values):")
print(top7_routes)

# 상위 7개 노선의 요일별 평균 지연 시간 출력
print("\nTop 7 Routes with Day-wise Delays:")
pivot_data_top7 = pivot_data.loc[top7_routes.index]
print(pivot_data_top7)

# Heatmap 시각화
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_data_top7, annot=True, fmt='.1f', cmap='magma', cbar_kws={'label': 'Average Delay (Minutes)'})
plt.title('Top 7 Origin-Destination Routes by Average Delay Across Days')
plt.xlabel('Day of the Week')
plt.ylabel('Origin-Destination Pair')
plt.show()

# 상위 7개 노선의 전체 평균 지연 시간 시각화
plt.figure(figsize=(10, 6))
top7_routes.plot(kind='bar', color='orange')
plt.title('Top 7 Origin-Destination Routes by Overall Average Delay')
plt.xlabel('Origin-Destination Pair')
plt.ylabel('Average Delay (Minutes)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
